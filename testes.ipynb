```python
import requests
import pandas as pd

API_KEY = "Q8D8PMFBZSPU99WQ35SKNWFBA"


def obter_previsao(local):
    url = f"https://weather.visualcrossing.com/VisualCrossingWebServices/rest/services/timeline/{local}/?key={API_KEY}"

    response = requests.get(url)

    if response.status_code == 200:
        dados = response.json()
        current_conditions = dados.get("currentConditions", {})
        resolved_address = dados.get("resolvedAddress", "N/A")
        description = dados.get("description", "N/A")
        df = pd.DataFrame.from_dict(current_conditions, orient="index")
        df = df.transpose()
        df_extra = pd.DataFrame(
            {"resolvedAddress": [resolved_address], "description": [description]}
        )
        df_final = pd.concat([df_extra, df], axis=1)

        return df_final

    else:
        print(f"Erro ao fazer a requisição: {response.status_code}")
        return None


df = obter_previsao("Santos, São Paulo")
```

```python
def convert_celsius(temp):
    temp = (temp - 32) * 5.0/9.0
    return round(temp)
```

```python
df["temp"] = df["temp"].apply(convert_celsius)
df["feelslike"] = df["feelslike"].apply(convert_celsius)
df["dew"] = df["dew"].apply(convert_celsius)

df
```
